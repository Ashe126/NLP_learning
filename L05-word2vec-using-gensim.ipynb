```python
database = '../L02languageModel/sqlResult_1558435.csv'
```

```python
import os
os.path.exists(database)  
```

```python
import pandas as pd
content = pd.read_csv(database, encoding='gb18030')
```

```python
samples = content['content'][:100].tolist()
```

```python
import jieba
def cut(string): return ' '.join(jieba.cut(string))
```

```python
cut('这是一个测试')
```

```python
with open('mini_samples.txt', 'w', encoding='utf-8') as f:
    for s in samples: 
        f.write(cut(s) + '\n')
```

```python
sentences = LineSentence('mini_samples.txt')
```

```python
model = Word2Vec(sentences, min_count=1)
```

```python
model.wv['小米']
```

```python
model.most_similar('小米')
```

```python
model.wv.vocab
```
